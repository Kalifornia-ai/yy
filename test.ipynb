```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

def plot_h5_dataset(h5_path, dataset_name):
    with h5py.File(h5_path, 'r') as f:
        data = f[dataset_name][:]
        
        # Check if data is complex
        if np.iscomplexobj(data):
            # Choose how to handle the complex data.
            # For example, let's plot the magnitude:
            data_to_plot = data[2]
            print(data_to_plot.shape)
        else:
            data_to_plot = data
        
        if 1 == 1:
            plt.figure()
            #plt.plot(data_to_plot)
            plt.title(f'{dataset_name} (1D)')
            plt.plot(data_to_plot[50000:50400])
            #plt.plot(np.abs(data_to_plot))
            plt.title("First 2k samples")
            plt.xlabel('Index')
            plt.ylabel('Value')
            plt.show()
        elif data.ndim == 2:
            plt.figure()
            factor = 10
            #data_downsampled = data_to_plot[:, :factor]  # keep every 10th column
            plt.imshow(data_to_plot, aspect='auto', cmap='viridis')
            plt.title(f'{dataset_name} (2D)')
            plt.colorbar()
            plt.show()
        else:
            print(f"{dataset_name} has shape {data.shape}. This script only plots 1D or 2D data.")

if __name__ == "__main__":
    h5_file_path = "./dataset/interferenceset_frame/EMISignal1_raw_data.h5"
    
    # Open to see what datasets are available:
    with h5py.File(h5_file_path, 'r') as f:
        print("Available datasets/groups:", list(f.keys()))
    
    
    
    # Suppose the dataset you want to plot is "dataset1"
    plot_h5_dataset(h5_file_path, 'dataset')
```
